```python
def multiple_return_greet():
    return 'Hello', 'World'


greet, name = multiple_return_greet()

print(greet)
print(name)
```

```python
def variable_arg_greet(*names):
    for name in names:
        print(f'hola {name}')

variable_arg_greet('Juan', 'Pedro', 'Maria')
```

```python
def variable_key_arg_greet(**names):
    for key, value in names.items():
        print(f'{key} {value}')


variable_key_arg_greet(
    lenguage='español',
    name='Juan',
    age=30,
    alias='Juancho'
)
```

```python
def print_numbers(text_1, text_2) -> int:
    for number in range(1, 101):
        if number % 3 == 0 and number % 5 == 0:
            print(text_1 + text_2)
        elif number % 3 == 0:
            print(text_1)
        elif number % 5 == 0:
            print(text_2)
        else:
            print(number)


print(print_numbers("Fizz", "Buzz"))
```

```python
def parametros(parametro_1, parametro_2): 
    contador = 0
    for number in range(1,101):
        if number % 3 != 0:
            print(parametro_1)
        elif number % 5 == 0:
            print(parametro_2)
        elif number % 3 == 0 and number % 5 == 0:
            print(parametro_1 + parametro_2)
        else:
            print(number)
            contador += 1
    return contador

print(parametros("Fizz", "Buzz"))
```
