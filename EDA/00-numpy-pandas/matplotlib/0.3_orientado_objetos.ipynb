Hay distintas maneras de hacer grficas dentro de Matplotlib, ya vimos pyplot; sin embargo, es muy complicado personalizarla y entrar a ciertos parmetros.

**Mtodo orientado a objetos en Matplotlib**


Un objeto define una figura, esa figura es un lienzo en el cual podemos introducir diferentes gr ficas(axes), de las cuales cada una posee sus propios ejes(axis).

La figura representa el todo, dentro de ella vamos a configurar grficas las cuales contienen diferentes ejes.

Es un poco ms complicado, pero en el mismo grfico podemos personalizarlo mucho mejor.



**Diferencias entre Pyplot y Object Oriented**

| Caracterstica          | Pyplot                 | Orientado a Objetos            |
| ----------------------- | ---------------------- | ------------------------------ |
| Velocidad               | Rpido                 | Mayor personalizacin         |
| Facilidad de Uso        | Fcil                  | Ms cdigo                     |
| Figura nica            | Una sola figura        | Ms amigable a mltiples diagramas |

## **Implementar el mtodo orientado a objetos para graficar en Matplotlib**

```python
x=np.linspace(0,5,11)
y=x**2
```

Parmetros de axes:
| Posicin Eje x | Posicin Eje y | Tamao Grfico Eje x | Tamao Grfico Eje y |
| --------------- | --------------- | -------------------- | -------------------- |

* Utilizamos el concepto de objeto para crear una grfica ms personalizable


```python
fig=plt.figure() #create a figure object

axes=fig.add_axes([0.1,0.1,0.5,0.9]) #left, bottom, width, height (range 0 to 1)
axes.plot(x,y,'b');
```

Podemos crear varias grficas dentro del lienzo llamando el mtodo fig.

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.9])
axes2=fig.add_axes([0.17,0.55,0.4,0.3]) # inset axes, we can create multiple axes in a figure

axes.plot(x,y,'b')
axes2.plot(y,x,'r')
axes2.set_facecolor('grey'); # change the background color of the inset axes, there are many other options to change the background color for example: axes2.set_facecolor('lightgrey')
```
