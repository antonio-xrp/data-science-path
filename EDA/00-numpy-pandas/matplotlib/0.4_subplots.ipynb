```python
import matplotlib.pyplot as plt
import numpy as np
```

## **Subplots**

Con subplots se puede trabajar en un arreglo de gráficas a las cuales se accede a través de los índices.

**Cómo trabajar con subplots para graficar**
* Creamos los datos que vamos a utilizar

```python
x = np.linspace(0,5,11)
y = x**2
```

**Crear una gráfica tipo Fig**


* Para crear una gráfica que sea tipo Fig. necesitas trabajar el siguiente código:

```python
fig,axes=plt.subplots() # we use the subplots method to create a figure and axes, (2,3) means 2 rows and 3 columns
axes.plot(x,y);
```

* Varias gráficas tipo fig.


```python
fig,axes=plt.subplots(nrows=1,ncols=2) 
axes[0].plot(x,y,"b") # we can use the axes object to plot on the specific subplot
axes[1].plot(y,x,"r");
```

Acceder a los gráficos mediante el nombre almacenado en una tupla.

```python
fig,(axes1,axes2)=plt.subplots(nrows=1,ncols=2) # we can use the axes object to plot on the specific subplot
axes1.plot(x,y,"b")
axes2.plot(y,x,"r");
```

Generar una matriz de tipo fig, con dos filas y cuatro columnas

```python
fig,axes=plt.subplots(2,4) 

axes[0,0].plot(x,y,"b") # "b" is the color blue
axes[0,1].plot(y,x,"r")
axes[1,0].plot(x,y,"b")
axes[1,1].plot(y,x,"r")
axes[0,2].plot(x,y,"b")
axes[0,3].plot(y,x,"r")
axes[1,2].plot(x,y,"b")
axes[1,3].plot(y,x,"r")

fig.tight_layout()
```

Alternativa para generar una matriz de tipo fig, accediendo a cada gráfica mediante los nombres dados en una tupla

```python
fig,((axes1,axes2,axes3,axes4),(axes5,axes6,axes7,axes8))=plt.subplots(2,4) # we can use the axes object to plot on the specific subplot
axes1.plot(x,np.sin(x),"b")
axes2.plot(x,np.cos(x),"purple")
axes3.plot(x,np.tan(x),"orange")
axes4.plot(x,np.tanh(x),"black")
axes5.plot(x,np.sin(x),"blue")
axes6.plot(x,np.cos(x),"purple")

fig.tight_layout()
```
