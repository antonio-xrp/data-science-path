Existen otros tipos de gráficos que Matplotlib nos proporciona para ser mucho más certeros en nuestros análisis.

**Crear otros tipos de gráficos en Matplotlib**

* Datos

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=np.random.randint(1,50,100);
```

Crear un gráfico tipo histograma


```python
plt.hist(data, bins=10, color='green'); # bins is the number of bars
```

Número de barras con Bins


```python
plt.hist(data, bins=50);
```

* Cambiar estilo


```python
plt.hist(data,bins=10,histtype="step"); # histtype="step" will draw a line instead of bars
```

Crear un gráfico tipo Boxplot


```python
plt.boxplot(data);
# #Muestra rango intercuartil y la mediana
```

Cambiar dirección, rellenar el rango inter cuartil patch_artist y enfocar la mediana notch

```python
plt.boxplot(data,vert=False,patch_artist=True,notch=True); # we use patch_artist to fill the boxplot with color and notch to draw a notch at the median, and we use notch to draw a notch at the median
```

* Quitar Outliers Showfliers


```python
data = np.append(data, 200)
plt.boxplot(data, vert=False, patch_artist=True, notch=True, showfliers=False)  # we use showfliers to hide the outliers
plt.show()
```

Crear un gráfico tipo Scatter

* Cargamos Datos


```python
N=50
x=np.random.rand(N)
y=np.random.rand(N)
area=(30*np.random.rand(N))**2
colors=np.random.rand(N)
```

* Scatter con Área


```python
plt.scatter(x,y,s=area); # s is the size of the dots
```

* Color y marker

```python
plt.scatter(x,y,s=area,c=colors,marker="v"); #marker is the shape of the dots
```

Transparencia para ver los superpuestos


```python
plt.scatter(x,y,s=area,c=colors,marker="o",alpha=0.5)
```
