```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Seaborn tiene una gran variedad de gráficos, pero también tiene ciertos parámetros para cada gráfico, vamos a ver cuáles son los más comunes:

## Parámetros más comunes al usar Seaborn

Cada uno de los gráficos utiliza ciertos parámetros que tu podrás gestionar. Empieza importando las librerías y el Dataset que vas a usar.

```python
tips=sns.load_dataset('tips') # we can use this dataset to practice with seaborn
tips.head()
```

* Gráfico de frecuencia para el total de la cuenta

| Tipo de gráfica | Dataset | X          |
|-----------------|---------|------------|
| .displot        | tips    | 'total_bill'|

## Pasos para usar los parámetros


```python
sns.displot(data=tips, x="total_bill"); # we can use displot to create a histogram, data is the dataset, x is the column we want to plot
plt.show()
```

* Agregamos una variable ‘Y’, pero seaborn reconoce que en gráfico de frecuencias no es la mejor opción y lo adapta automáticamente.

```python
sns.displot(data=tips, x="total_bill", y="tip"); # we use x and y to create a scatter plot
```

* Hacer agrupamiento por una variable hue


```python
sns.displot(data=tips, x="total_bill", y="tip", hue="sex"); # we use hue to add a third dimension to the plot by using one of the columns as a color
```

* Cambiar el tipo de gráfica kind


```python
sns.displot(data=tips,x="total_bill", hue="sex", kind="kde"); # we can use kind to change the type of plot, in this case we use kde to create a density plot, also there are other options like hist, ecdf, and more
```

* Quitar la legenda, cambiar la paleta de colores y su transparencia

```python
sns.displot(data=tips,x="total_bill",hue="sex",kind="kde", legend=True, palette="colorblind", alpha=.5); #we use legente to show the legend, palette to change the color, and alpha to change the transparency. There other options to change the color, like "deep", "muted", "pastel", "bright", "dark", and "colorblind"
```
