```python
import pandas as pd
import numpy as np
```

Muchas ocasiones necesitamos agregar, eliminar o separar datos y pandas nos ofrece varias funciones para que este proceso se vuelva mucho más sencillo.

* Muestra las primeras 5 líneas del DataFrame

```python
df_books = pd.read_csv('data_example.csv')
```

```python
df_books.head() # Show the first 5 rows
```

* Eliminar columnas de la salida pero no del DataFrame


#elimina la columna Genre de la salida pero no del dataFrame

```python
df_books.drop("Genre", axis=1).head() # Drop the column Genre, we use axis=1 to indicate that we are dropping a column
```

* Eliminar una columna


#elimina la columna Price del dataFrame

```python
del df_books["Price"] # Another way to drop a column using the del statement
```

* Eliminar filas

#elimina la fila 0 del dataFrame

```python
df_books.drop(0,axis=0).head() # Drop the first row, we use axis=0 to indicate that we are dropping a row
```

* Eliminar un conjunto de filas mediante una lista


#elimina las filas 0, 1 y 2 del dataFramem

```python
df_books.drop([0,1,2],axis=0).head() # Drop the first three rows, we use axis=0 to indicate that we are dropping a row
```

* Elimina un conjunto de filas mediante un rango

 #elimina las primeras 10 filas del dataFrame

```python
df_books.drop(range(0,10),axis=0).head() # Drop the first ten rows, we use range to indicate the rows we want to drop
```

* Agregar una nueva columna con valores Nan

#Crea una nueva columna con el nombre de Nueva_columna de valores Nan

```python
#df_books["Nueva_columna"]=np.nan
df_books["Nueva_columna"]=pd.NA # Add a new column with NaN values, we use pd.NA to indicate that we want to add NaN values

df_books.head()
```

Mostrar el número de filas o columnas que tiene un DataFrame


#Muestra el numero de filas que posee el dataFrame

```python
df_books.shape[0] # Get the number of rows 
```

* Agregar valores a una nueva columna


```python
data=np.arange(0,df_books.shape[0]) # Create an array with the same number of rows as the dataframe, we use np.arange to create the array
data
```

* Crear una nueva columna y agregar los valores almacenados en el array

```python
df_books["Rango"]=data # Add the array to the dataframe as a new column
df_books.head()
```

* Para añadir filas se utiliza la función append de Python añadiendo como parámetro una lista, diccionario o añadiendo los valores manualmente.

```python
new_row ={"Name":"New book","Author":"New author","User Rating":4.5,"Reviews":1000,"Year":2020,"Nueva_columna":1}

df_books=pd.concat([df_books,pd.DataFrame([new_row])],ignore_index=True) # Add a new row to the dataframe, we use pd.concat to add the new row, we use ignore_index=True to reset the index of the dataframe

df_books
```
