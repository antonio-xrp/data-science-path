# Regresin logstica

La regresin logstica es una tcnica de anlisis estadstico que se utiliza para predecir el resultado de una variable categrica (como "s" o "no", "verdadero" o "falso", etc.) basada en una o ms variables independientes (que pueden ser numricas o categricas). A diferencia de la regresin lineal, que predice valores continuos, la regresin logstica se utiliza para predecir la probabilidad de que un evento ocurra.
