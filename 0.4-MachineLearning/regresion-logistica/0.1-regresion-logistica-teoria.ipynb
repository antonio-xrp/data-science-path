# Regresión logística

La regresión logística es una técnica de análisis estadístico que se utiliza para predecir el resultado de una variable categórica (como "sí" o "no", "verdadero" o "falso", etc.) basada en una o más variables independientes (que pueden ser numéricas o categóricas). A diferencia de la regresión lineal, que predice valores continuos, la regresión logística se utiliza para predecir la probabilidad de que un evento ocurra.
