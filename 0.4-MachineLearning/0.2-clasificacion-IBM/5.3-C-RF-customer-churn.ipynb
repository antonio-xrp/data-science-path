<h2 id="TargPreproc">Part 2: Examining the Target and Preprocessing</h2>

In this exercise, we will proceed as follows:

*   Examine distribution of the predicted variable (`churn_value`).
*   Split the data into train and test sets. Decide if a stratified split should be used or not based on the distribution.
*   Examine the distribution of the predictor variable in the train and test data.


```python
# Data are skewed at ~85% towards non-churned customers
# This will be important to remember when model building
target = 'churn_value'
data[target].value_counts()
```

```python
data[target].value_counts(normalize=True)
```

Given the skew in the predictor variable, let's split the data with the *churned* values being stratified to ensure that the test and train sets have similar distributions.


```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
feature_cols = [x for x in data.columns if x != target]
```

```python
# Split the data into two parts with 1500 points in the test data
# This creates a generator
strat_shuff_split = StratifiedShuffleSplit(n_splits=1, test_size=1500, random_state=42)
```

```python
# Get the index values from the generator
train_idx, test_idx = next(strat_shuff_split.split(data[feature_cols], data[target]))
```

```python
# Create the data sets
X_train = data.loc[train_idx, feature_cols]
y_train = data.loc[train_idx, target]

X_test = data.loc[test_idx, feature_cols]
y_test = data.loc[test_idx, target]
```

Let's check the distributions:


```python
y_test.value_counts(normalize=True)
```

<h2 id ="OOB">Part 3: Random Forest and Out-of-bag(oob) Error</h2>

In this exercise, we will:

*   Fit random forest models with a range of tree numbers and evaluate the out-of-bag error for each of these models.
*   Plot the resulting oob errors as a function of the number of trees.

*Note:* since the only thing changing is the number of trees, the `warm_start` flag can be used so that the model just adds more trees to the existing model each time. Use the `set_params` method to update the number of trees.


```python

# Suppress warnings about too few trees from the early models
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
warnings.filterwarnings("ignore", category=RuntimeWarning)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

El parÃ¡metro `oob_score` (out-of-bag score) en el clasificador `RandomForestClassifier` se utiliza para medir el rendimiento del modelo a partir de las muestras _"out-of-bag" (fuera de la bolsa)_. En el contexto de los bosques aleatorios (Random Forests), cada Ã¡rbol del bosque se entrena con una muestra de los datos obtenida mediante _"bootstrap sampling"_, es decir, una selecciÃ³n aleatoria con reemplazo. Como resultado, algunas observaciones no se incluyen en la muestra de entrenamiento de un Ã¡rbol especÃ­fico. Estas observaciones excluidas se denominan "out-of-bag" (fuera de la bolsa).

El `oob_score=True` permite que el modelo use estas muestras fuera de la bolsa para estimar el rendimiento del modelo sin necesidad de realizar una validaciÃ³n cruzada o separar un conjunto de validaciÃ³n. El cÃ¡lculo del `oob_score`es Ãºtil porque es una forma eficiente de obtener una medida del error de generalizaciÃ³n del modelo mientras se entrena.

```python
# Initialize the random forest estimator
# Note that the number of trees is not setup here
RF = RandomForestClassifier(oob_score=True,
                            random_state=42,
                            warm_start=True,
                            n_jobs=1)
```

```python
oob_list = list()
```

```python
# Iterate through all of the possibilities for 
# number of trees

for n_trees in [15, 20, 30, 40, 50, 100, 150, 200, 300, 400]:
     # Use this to set the number of trees
    RF.set_params(n_estimators=n_trees)

    # Fit the model
    RF.fit(X_train, y_train)

    # Get the oob error
    oob_error = 1 - RF.oob_score_
    
    # Store it
    oob_list.append(pd.Series({'n_trees': n_trees, 
                               'oob': oob_error}))
    
rf_oob_df = pd.concat(oob_list, axis=1).T.set_index('n_trees')

rf_oob_df
```

The error looks like it has stabilized around 100-150 trees.


El OOB error mide quÃ© tan bien el modelo predice los datos que no fueron utilizados para entrenar cada Ã¡rbol en particular. Mientras menor sea este valor, mejor es la capacidad predictiva del modelo.

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
sns.set_context('talk')
sns.set_style('white')

ax = rf_oob_df.plot(legend=False, marker='o', figsize=(14, 7), linewidth=5)
ax.set(ylabel='out-of-bag error');
plt.show()
```

<h2 id = "ExtraTrees">Part 4: Extra Trees</h2>

Our exercise:

*   Repeat question 3 using extra randomized trees (`ExtraTreesClassifier`). Note that the `bootstrap` parameter will have to be set to `True` for this model.
*   Compare the out-of-bag errors for the two different types of models.


```python
from sklearn.ensemble import ExtraTreesClassifier
```

`warm_start=True:` Este parÃ¡metro permite al modelo reutilizar los Ã¡rboles que ya fueron entrenados en una ejecuciÃ³n anterior cuando el modelo se ajusta nuevamente. En lugar de empezar de cero cada vez que llamas al mÃ©todo fit, el modelo puede seguir aÃ±adiendo Ã¡rboles adicionales a los ya existentes.

Esto es Ãºtil cuando quieres ajustar un modelo de forma incremental, por ejemplo, si deseas probar el rendimiento con diferentes nÃºmeros de Ã¡rboles sin reconstruir el modelo desde cero cada vez. En tu caso, puedes usarlo para agregar mÃ¡s Ã¡rboles al bosque sin perder los que ya se entrenaron.

```python
# Initialize the random forest estimator
# Note that the number of trees is not setup here
EF = ExtraTreesClassifier(oob_score=True, 
                          random_state=42, 
                          warm_start=True,
                          bootstrap=True,
                          n_jobs=-1)
```

```python
oob_list = list()
```

```python
# Iterate through all of the possibilities for 
# number of trees

for n_trees in [15, 20, 30, 40, 50, 100, 150, 200, 300, 400]:

    EF.set_params(n_estimators = n_trees)
    EF.fit(X_train, y_train)


    oob_error = 1 - EF.oob_score_
    oob_list.append(pd.Series({'n_trees':n_trees,
                               'oob': oob_error}))

et_oob_df = pd.concat(oob_list, axis=1).T.set_index('n_trees')

et_oob_df
```

Combine the two dataframes into a single one for easier plotting.


```python
oob_df = pd.concat([rf_oob_df.rename(columns={'oob':'RandomForest'}),
                    et_oob_df.rename(columns={'oob':'ExtraTrees'})], axis=1)

oob_df
```

The random forest model performs consistently better than the extra randomized trees.


```python
sns.set_context('talk')
sns.set_style('white')

ax = oob_df.plot(marker='o', figsize=(14, 7), linewidth=5)
ax.set(ylabel='out-of-bag error');
plt.show()
```

<h2 id="results">Part 5: Gathering Results</h2>

Here, we will:

*   Select one of the models that performs well and calculate error metrics and a confusion matrix on the test data set.
*   Given the distribution of the predicted class, which metric is most important? Which could be deceiving?


```python
# Random forest with 100 estimators
model = RF.set_params(n_estimators = 100)

y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score
from sklearn.metrics import f1_score, roc_auc_score
```

Unsurprisingly, recall is rather poor for the customers who churned (True) class since they are quite small. We are doing better than random guessing, though, as the accuracy is 0.94 (vs 0.85 for random guessing).


```python
cr = classification_report(y_test, y_pred)
print(cr)
```

```python
score_df = pd.DataFrame({'accuracy': accuracy_score(y_test, y_pred),
                         'precision': precision_score(y_test, y_pred),
                         'recall': recall_score(y_test, y_pred),
                         'f1': f1_score(y_test, y_pred),
                         'auc': roc_auc_score(y_test, y_pred)},
                         index=pd.Index([0]))

print(score_df)
```

<h2 id="examineRes">Part 6: Examining Results</h2>

The following exercises will help us examine results:

*   Print or visualize the confusion matrix.
*   Plot the ROC-AUC and precision-recall curves.
*   Plot the feature importances.


```python
from sklearn.metrics import roc_curve, precision_recall_curve, confusion_matrix,ConfusionMatrixDisplay
```

```python
sns.set_context('talk')
cm = confusion_matrix(y_test, y_pred,normalize='true')
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=model.classes_)
disp.plot()
plt.show()
```

98% TF no se fueron, el 2% si fueron pero clasifica como que no se fueron. 
el 83% se fueron, el 17 se fueron pero clasifca como que no se fueron. 

The ROC-AUC and precision-recall curves.


```python
sns.set_context('talk')

fig, axList = plt.subplots(ncols=2)
fig.set_size_inches(16, 8)
```

```python
# Get the probabilities for each of the two categories
y_prob = model.predict_proba(X_test)
```

```python
# Plot the ROC-AUC curve
ax = axList[0]
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_prob[:,1])
ax.plot(fpr, tpr, linewidth=5)
# It is customary to draw a diagonal dotted line in ROC plots.
# This is to indicate completely random prediction. Deviation from this
# dotted line towards the upper left corner signifies the power of the model.
ax.plot([0, 1], [0, 1], ls='--', color='black', lw=.3)
ax.set(xlabel='False Positive Rate',
       ylabel='True Positive Rate',
       xlim=[-.01, 1.01], ylim=[-.01, 1.01],
       title='ROC curve')
ax.grid(True)

# Plot the precision-recall curve
ax = axList[1]

precision, recall, _ = precision_recall_curve(y_test, y_prob[:,1])
ax.plot(recall, precision, linewidth=5)
ax.set(xlabel='Recall', ylabel='Precision',
       xlim=[-.01, 1.01], ylim=[-.01, 1.01],
       title='Precision-Recall curve')
ax.grid(True)

plt.tight_layout()
plt.show()
```

**ROC**

- La curva azul estÃ¡ claramente por encima de la lÃ­nea de referencia, lo que indica que el modelo tiene un buen rendimiento en la distinciÃ³n entre clases.
- Mientras mÃ¡s cerca estÃ© la curva de la esquina superior izquierda, mejor es el rendimiento del modelo, lo que sugiere que el modelo tiene una buena capacidad predictiva.

**Precision-recall**

- Este grÃ¡fico es especialmente Ãºtil cuando hay un desequilibrio entre las clases (cuando una clase es mucho mÃ¡s comÃºn que la otra).
- La curva muestra un rendimiento bastante bueno: alta precisiÃ³n y alta sensibilidad en la mayorÃ­a de los umbrales.
- Hacia la derecha del grÃ¡fico (recall cercano a 1), la precisiÃ³n disminuye, lo que es comÃºn cuando intentas clasificar todos los positivos posibles (recall alto), sacrificando precisiÃ³n al incluir mÃ¡s falsos positivos.
