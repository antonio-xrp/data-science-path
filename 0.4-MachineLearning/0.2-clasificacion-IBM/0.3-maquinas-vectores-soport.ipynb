### Implementing Faster Kernel Transformations 



```python
# Import the class containing the classification method
from sklearn.kernel_approximation import Nystroem

# Crate an isntance of the class
NystroemSVC = Nystroem(kernel='rbf', gamma=1.0, n_components=100) # n_components is number of samples

# fit the instance on the data and transform 

X_train = NystroemSVC.fit_transform(X_train)
X_test = NystroemSVC.transform(X_test)

Tune kernel and associated parameters with cross-validation

# Import the class containing the classification method
from sklearn.kernel_approximation import RBFsampler

# Create an instance of the class
rbfSample = RBFsampler(gamma=1.0, n_components=100)
# RBF is only kernel that can be used 
# parameters names are indentical to previous

# Fit the instance on the data and transform
X_train = rbfSample.fittransform(X_train)
X_test = rbfSample.transform(X_test)

Tune kernel and associated parameters with cross-validation
```
