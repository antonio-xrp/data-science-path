### Stationarity and Autocorrelation

**Stationarity** impacts our ability to model and forecast

- Una serie es estacionaria si tiene la misma media y varianza a lo largo del tiempo, lo que facilita su modelado.
- Para manejar series no estacionarias, primero se identifica la fuente de no estacionariedad y luego se aplica una transformacin para hacerla estacionaria.

- A **stationary** series has the same mean and variance over time
- **Non-stationary** series are much harder to model

Common approach:
- Indentify sources of non-stationarity
- Transform series to make it stationary
- Build models with stationary series

### Assessing Stationarity
- **Media constante:** La media de la serie debe permanecer constante; un ejemplo de no estacionariedad sera una serie con tendencia ascendente o descendente.

- **Varianza constante:** La varianza debe ser constante; un ejemplo de no estacionariedad sera un portafolio de inversin que se vuelve voltil tras un perodo de estabilidad.
