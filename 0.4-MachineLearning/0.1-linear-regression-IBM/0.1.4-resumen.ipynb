**Introducción al aprendizaje automático supervisado**  

Los tipos de aprendizaje automático supervisado son

- Regresión, en el que la variable objetivo es continua

- Clasificación, en la que la variable objetivo es categórica

Para construir un modelo de clasificación se necesitan

- Características que puedan cuantificarse

- Una variable objetivo o de resultado etiquetada

- Un método para medir la similitud

### Regresión lineal  
Una regresión lineal modela la relación entre una variable continua y una o más variables de escala.Suele representarse como una función dependiente igual a la suma de un coeficiente más los factores de escala multiplicados por las variables independientes.

Los residuos se definen como la diferencia entre un valor real y un valor predicho.

Una práctica recomendada de modelización para la regresión lineal es:

- Utilizar la función de coste para ajustar el modelo de regresión lineal

- Desarrolle múltiples modelos

- Compare los resultados y elija el que se ajuste a sus datos y a si está utilizando su modelo para la predicción o la interpretación.

Tres medidas comunes de error para regresiones lineales son:

- Suma de errores al cuadrado (SSE)

- Suma total de cuadrados (SST)

- Coeficiente de determinación (R2)
