### Modelos unsupervisados 

## K-means: Modelo de Aprendizaje No Supervisado

**K-means** es un algoritmo de clustering (agrupamiento) no supervisado que se utiliza para agrupar datos en diferentes "clusters" (grupos) basados en sus características. Es útil para identificar patrones o segmentar datos sin etiquetas o categorías predefinidas.

### Concepto Sencillo:

- Imagina que tienes un conjunto de puntos en un plano y quieres agruparlos en diferentes grupos basándote en su proximidad.
- K-means agrupa los puntos asignándolos al grupo cuyo centro (centroide) esté más cercano.
- Inicialmente, se eligen aleatoriamente algunos puntos como centros de los grupos (en este caso, 3 centros).
- Luego, cada punto se asigna al centro más cercano.
- El algoritmo recalcula los centros como el promedio de los puntos asignados a cada grupo y repite el proceso hasta que los centros ya no cambian mucho.

- En el gráfico, los puntos están distribuidos en tres grupos.
- Los colores indican a qué grupo pertenece cada punto.
- Las marcas rojas ('X') son los centroides de cada grupo, que representan los centros calculados por el algoritmo.

Este es un ejemplo básico de cómo K-means organiza datos en grupos basados únicamente en sus posiciones relativas, sin necesidad de etiquetas previas.
