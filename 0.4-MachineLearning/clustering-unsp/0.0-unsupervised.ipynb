## K-means: Modelo de Aprendizaje No Supervisado

**K-means** es un algoritmo de clustering (agrupamiento) no supervisado que se utiliza para agrupar datos en diferentes "clusters" (grupos) basados en sus caractersticas. Es til para identificar patrones o segmentar datos sin etiquetas o categoras predefinidas.

### Concepto Sencillo:

- Imagina que tienes un conjunto de puntos en un plano y quieres agruparlos en diferentes grupos basndote en su proximidad.
- K-means agrupa los puntos asignndolos al grupo cuyo centro (centroide) est ms cercano.
- Inicialmente, se eligen aleatoriamente algunos puntos como centros de los grupos (en este caso, 3 centros).
- Luego, cada punto se asigna al centro ms cercano.
- El algoritmo recalcula los centros como el promedio de los puntos asignados a cada grupo y repite el proceso hasta que los centros ya no cambian mucho.
