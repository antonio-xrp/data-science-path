### How to train a neural net?

La retropropagacin es como un maestro que corrige un examen. Imagina que tienes un grupo de estudiantes (que representan las neuronas de una red) que intentan resolver un problema. Al final, el maestro (la funcin de prdida) revisa las respuestas y les dice a los estudiantes qu tan bien lo hicieron. Si un estudiante se equivoc, el maestro le da consejos sobre cmo mejorar.

Ahora, en lugar de solo decirles que se equivocaron, el maestro les dice exactamente qu parte de su respuesta fue incorrecta y cmo pueden ajustar su enfoque para hacerlo mejor la prxima vez. Esto es lo que hace la retropropagacin: ayuda a cada "neurona" a entender cmo debe cambiar su "respuesta" (o peso) para que el grupo en su conjunto obtenga una mejor calificacin.

**Ejemplo Prctico**

Imagina que ests aprendiendo a lanzar una pelota a un cesto. Al principio, lanzas la pelota y no entra. Luego, alguien te dice que debes lanzar un poco ms alto. La prxima vez que lances, ajustas tu tcnica basndote en ese consejo. Cada vez que lanzas, recibes retroalimentacin y haces pequeos ajustes hasta que finalmente logras encestar. As funciona la retropropagacin: ajustamos los pesos de la red neuronal basndonos en el error que cometimos, para mejorar nuestras predicciones.

### The sigmoid Activation Function

Las funciones de activacin son como interruptores que deciden si una neurona en una red neuronal debe "encenderse" o "apagarse". Imagina que ests en una habitacin oscura y tienes un interruptor de luz. Cuando enciendes el interruptor, la luz se enciende y puedes ver todo a tu alrededor. De manera similar, las funciones de activacin permiten que las redes neuronales tomen decisiones sobre qu informacin es importante y cul no. Por ejemplo, la funcin sigmoide, que hemos mencionado, convierte cualquier nmero en un valor entre 0 y 1, lo que es til para problemas donde queremos resultados claros, como clasificar algo como "s" o "no".

Sin embargo, hay un pequeo problema con la funcin sigmoide: a veces, cuando los nmeros son muy grandes o muy pequeos, la luz apenas se enciende, lo que significa que la red tiene dificultades para aprender. Esto se llama el "problema del gradiente que se desvanece". As que, aunque la funcin sigmoide es til, hay otras funciones de activacin que pueden ayudar a las redes neuronales a aprender mejor y ms rpido.

### Other Popular activation functions

**La funcin ReLU**: es como un filtro que decide qu informacin dejar pasar en una red neuronal. Si el valor de entrada (llamado "z") es mayor que cero, simplemente lo deja pasar tal cual. Pero si el valor es menor o igual a cero, lo convierte en cero. Imagina que tienes una puerta: si alguien est de pie (valor positivo), la puerta se abre y entra. Si est sentado o no hay nadie (valor cero o negativo), la puerta permanece cerrada. Esto ayuda a que la red neuronal se enfoque en las partes ms importantes de la informacin.


**Variante "Leaky ReLU":** . En lugar de cerrar la puerta completamente cuando el valor es negativo, deja que una pequea cantidad de informacin pase. As, si el valor es negativo, lo multiplica por un nmero pequeo (como 0.01) en lugar de convertirlo en cero. Esto es til porque asegura que la red siga aprendiendo, incluso cuando algunos valores son negativos.

### coded summary

```python
# Import libraries, model  elements 
from keras.models import Sequential
from keras.layers import Dense, Activation

model = Sequential()

# For the first layer, specify the input dimension
model.add(Dense(units=4, input_dim=3))

# Specify activation function
model.add(Activation('sigmoid'))

# For subsequent layers, the input dimension is presumed from the previous layer
model.add(Dense(units=4))
model.add(Activation('sigmoid'))
```
