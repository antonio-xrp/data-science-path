### Forward propagation 

Imagina que ests en una cocina preparando una receta. **La capa de entrada** sera como los ingredientes que tienes en la mesa, como la harina, el azcar y los huevos. Cada uno de estos ingredientes representa una caracterstica de tus datos. Luego, tienes las capas ocultas, que son como los pasos de la receta donde mezclas y cocinas esos ingredientes. Cada paso transforma los ingredientes de alguna manera, y en una red neuronal, esto se hace a travs de conexiones llamadas pesos. Estos pesos determinan cunto de cada ingrediente (o caracterstica) se usa en cada paso.

Finalmente, llegas a la **capa de salida**, que es como el plato terminado. Aqu es donde ves el resultado final de tu receta, que en el caso de una red neuronal, son las predicciones que hace el modelo. As que, en resumen, las capas en una red neuronal trabajan juntas para transformar los datos de entrada en resultados tiles, igual que en una buena receta!



### Matrix representation of forward propagation

Entendiendo el Proceso de una Red Neuronal de Manera Sencilla

Imagina que una red neuronal es como un grupo de amigos que estn tratando de resolver un rompecabezas. Cada amigo tiene una tarea especfica y, al final, todos trabajan juntos para encontrar la solucin.

1. **Recibiendo la Informacin:** Primero, uno de los amigos recibe un conjunto de datos, como una lista de caractersticas de algo que quieren clasificar (por ejemplo, diferentes tipos de frutas). Cada caracterstica es como una pista que les ayudar a resolver el rompecabezas.

2. **Multiplicacin y Decisin:** Este amigo pasa la informacin a otro amigo, que tiene una serie de "pesos" (que son como las prioridades que cada pista tiene). Este amigo multiplica las pistas por esos pesos para obtener un nuevo conjunto de valores, que llamamos Z1. Luego, aplica una funcin de activacin, que es como un filtro que decide si esos valores son tiles o no. Es como si dijera: "Hmm, esta pista es importante, pero esta otra no tanto".

3. **Pasando la Informacin:** Luego, el resultado (A1) se pasa a otro amigo, que hace lo mismo: multiplica por nuevos pesos y aplica otro filtro. Este proceso se repite varias veces, cada vez refinando la informacin hasta que todos los amigos han hecho su parte.

4. **Llegando a una Conclusin:** Al final, el ltimo amigo toma todos esos resultados y los convierte en probabilidades. Esto les dice qu tan seguros estn de que la fruta es, por ejemplo, una manzana, una naranja o un pltano.
