### Basic of neurons 
L**a Neurona en una Red Neuronal: Una Explicacin Simple**
Imagina que una neurona en una red neuronal es como un chef en una cocina. Este chef recibe varios ingredientes (que son los valores de entrada) y los mezcla en una olla (que representa la combinacin de esos valores). Cada ingrediente tiene una importancia diferente, as que el chef ajusta la cantidad de cada uno (esto se hace a travs de los pesos). Despus de mezclar, el chef tiene que decidir si la mezcla es buena o no, y para eso utiliza una receta especial (la funcin de activacin) que le ayuda a transformar la mezcla en un plato final que puede ser servido (el valor de salida).

**Ejemplo para Ilustrar**
Supongamos que el chef tiene tres ingredientes: zanahorias, cebollas y especias. Cada uno tiene un peso diferente segn lo que el chef prefiere. Despus de mezclar estos ingredientes, el chef usa su receta para decidir si el plato es delicioso (un valor entre 0 y 1) o no. Si el plato es delicioso, lo sirve a la siguiente cocina (la siguiente capa de la red neuronal). Sin esta receta, el chef solo podra hacer platos simples, pero con ella, puede crear sabores complejos y deliciosos.
