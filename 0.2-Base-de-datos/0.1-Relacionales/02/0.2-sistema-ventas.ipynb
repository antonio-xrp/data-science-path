- cp

- descripcion
- foto

- cantidad

### ventas **(ED)**

### artculos_x_venta **(EP)**

- pais_id **(pk)**
- nombre
- dominio **(UQ)**

## Relaciones

1. Un **cliente** tiene **pais** (_1 - 1_)
2. Un **cliente** genera **ventas** (_1 - M_)
2. Un **venta** tiene **articulos** (_1 - M_)
3. un **artculo** tiene un **producto** (_1 - 1_)
