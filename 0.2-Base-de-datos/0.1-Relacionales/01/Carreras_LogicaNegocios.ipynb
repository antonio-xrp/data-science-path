# Carreras 

## Listado de Entidades

### carreras **(ED)**
- carrera_id **(Pk)**
- nombre
- tipo_carrera **(FK)**
- fecha
- tiempo
- mejor_tiempo
- altitud
- lugar
- pais **(FK)**
- foto

### tipos_carreras **(EC)**
- tipos_carrera_id **(Pk)**
- descripcion
- distancia **(UK)**

### paises **(EC)**
- pais_id **(Pk)**
- nombre_pais

### Relaciones 
1. Una **carrera** _pertenece_ a un **tipo de carrera** (_1 a 1_)
2. Una **carrera** se _corre_ en un **país**. (_1 a 1_)
