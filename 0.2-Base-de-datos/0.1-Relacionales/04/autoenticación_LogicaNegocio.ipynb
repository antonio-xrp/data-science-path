# Sistema de Autenticación

## Listado de entidades 
--- 
### usuarios (ED)
- usuarios_id **(PK)**
- username **(UQ)**
- password 
- email **(UQ)**
- nombre
- apellido
- avatar 
- activo
- fecha_creacion
_ fecha_actualizacion

### roles (Ec)
- rol_id **(PK)**
- nombre
- descripcion

### permisos (EC)
- permiso_id **(PK)**
- nombre
- descripcion

### roles_x_usuarios (EP)
- rxu_id **(PK)**
- usuario_id **(FK)**
- rol_id **(FK)**

### permisos_x_roles (EP)
- pxr_id **(PK)**
- rol_id **(FK)**
- permiso_id  **(FK)**
---
### Relaciones 
1. usuarios tiene roles (M-M)
2. Roles tienen permisos (M-M)
