```python
import seaborn as sns
import matplotlib.pyplot as plt 
```

seaborn ofrece diferentes gráficos para graficar datos categóricos o variables de texto. A continuación verás el código que podrás usar para gráficar este tipo de datos.

## Tipos de datos categóricos para graficar en 


Empieza importando los datos que necesitas de las librerías

### Gráfica de cantidad


```python
tips = sns.load_dataset('tips')
tips.head(2)
```

```python
sns.countplot(data=tips,x="day", hue="sex"); # we use countplot to count the number of occurences of each category in a column
```

### Diagrama de puntos


```python
sns.swarmplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True) # we use swarmplot to show the distribution of a categorical variable across a second categorical variable
```

### Diagrama de caja


```python
sns.boxplot(data=tips, x="day",y="total_bill",hue="sex", dodge=True) # we use boxplot to show the distribution of a categorical variable across a second categorical variable
```

* Se pueden combinar distintos gráficos


```python
sns.boxplot(data=tips,x="day",y="total_bill", hue="sex",dodge=True);
sns.swarmplot(data=tips,x="day",y="total_bill",hue="sex", dodge=True,color="0", marker="<")
```

* Violín: parecido a Boxplot, pero no muestra los cuartiles, sino como se concentran los datos con comparación split

```python
sns.violinplot(data=tips,x="day",y="total_bill",hue="sex",split=True) # we use violinplot to show the distribution of a categorical variable across a second categorical variable
```

* Catplot, con kind podemos cambiar el tipo de gráfico y se puede agregar otra variable col


```python
sns.catplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True, kind="box", col="time") # we use catplot to show the distribution of a categorical variable across a second categorical variable, alsoe col argument is used to split the plot into two columns
```
