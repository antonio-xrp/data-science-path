## Merge y Concat


Como podemos usar la lógica anteriormente vista en código, usando los parámetros de Pandas

* Importamos Pandas y Numpy

**Concat**
* En esta ocasión vamos a crear un DataFrame nuevo


```python
df1 = pd.DataFrame({'A':['A0', 'A1', 'A2','A3'],
        'B':['B0', 'B1', 'B2','B3'],
	'C':['C0', 'C1', 'C2','C3'],
	'D':['D0', 'D1', 'D2','D3']})


df2 = pd.DataFrame({'A':['A4', 'A5', 'A6','A7'],
	'B':['B4', 'B5', 'B6','B7'],
	'C':['C4', 'C5', 'C6','C7'],
	'D':['D4', 'D5', 'D6','D7']})
```

```python
df1
```

```python
df2
```

Concatenar los DataFrames


```python
pd.concat([df1,df2]) #concatenating the dataframes
```

Corregir los índices


```python
pd.concat([df1,df2],ignore_index=True) #concatenating the dataframes and ignoring the index
```

Por axis 1


```python
pd.concat([df1,df2],axis=1) #concatenating the dataframes along the columns
```

## **Merge**

Creamos DataFrame

```python
izq = pd.DataFrame({'key' : ['k0', 'k1', 'k2','k3'],
 'A' : ['A0', 'A1', 'A2','A3'],
'B': ['B0', 'B1', 'B2','B3']})

der = pd.DataFrame({'key' : ['k0', 'k1', 'k2','k3'],
 'C' : ['C0', 'C1', 'C2','C3'],
'D': ['D0', 'D1', 'D2','D3']})
```

```python
izq
```

```python
der
```

Unir el DataFrame Der a Izq


```python
izq.merge(der) #merging the dataframes, the key is the column that is common in both dataframes
```

```python
#MERGE 2
izq = pd.DataFrame({'key' : ['k0', 'k1', 'k2','k3'],
 'A' : ['A0', 'A1', 'A2','A3'],
'B': ['B0', 'B1', 'B2','B3']})

der = pd.DataFrame({'key_2' : ['k0', 'k1', 'k2','k3'],
 'C' : ['C0', 'C1', 'C2','C3'],
'D': ['D0', 'D1', 'D2','D3']})
```

```python
izq
```

```python
der
```

Hay diferencias entre algunas columnas, por esa razón hay que separarlos de esta manera

```python
izq.merge(der, left_on="key", right_on="key_2") #merging the dataframes, the key is the column that is common in both dataframes, we use left_on and right_on to specify the column names
```

```python
# MERGE 3

izq = pd.DataFrame({'key' : ['k0', 'k1', 'k2','k3'],
 'A' : ['A0', 'A1', 'A2','A3'],
'B': ['B0', 'B1', 'B2','B3']})

der = pd.DataFrame({'key_2' : ['k0', 'k1', 'k2',np.nan],
 'C' : ['C0', 'C1', 'C2','C3'],
'D': ['D0', 'D1', 'D2','D3']})
```

```python
izq
```

```python
der
```

Si tenemos un NaNen nuestro DataFrame, pandas no lo detectará como un mach. Se soluciona con How, dando así, una preferencia.

```python
izq.merge(der, left_on="key", right_on="key_2", how="left") #merging the dataframes, the key is the column that is common in both dataframes, we use left_on and right_on to specify the column names, how="left" means that we are doing a left join
```
