Cuando queremos navegar por un dataFrame estas funciones permiten filtrar datos de manera más específica

## **.loc**


Filtra según un label

---> #muestra todos los datos del dataFrame


```python
df_books.loc[:] # all rows and columns
```

Mostrar un rango de filas tomando en cuenta el start y el end

---> #muestra los datos de la fila 0 a la fila 4




```python
df_books.loc[0:4] # first 5 rows and all columns
```

Filtrando por filas y columnas

 #filtra los datos de la fila que va de 0 a 4 y de las columnas Name y Autho


```python
df_books.loc[0:4,["Name","Author"]] # first 5 rows and 2 columns
```

* Podemos modificar los valores de una columna específica del dataFrame

#multiplica por -1 todos los valores de la columna Reviews

```python
df_books.loc[:,["Reviews"]]*-1 # all rows and 1 column with negative values
```

Filtrar datos que cumplan una condición determinada


```python
df_books.loc[:,["Author"]]=="JJ Smith" # all rows and 1 column with boolean values, True if Author is JJ Smith
```

## **.iloc**
Filtra mediante índices.

 #muestra todos los datos del dataframe

```python
df_books.iloc[:] # all rows and columns
```

Filtrar datos según los índices de las filas y las columnas

#muestra los datos de las filas que van de 0 a 3 y las columnas con indices 0 y 1

```python
df_books.iloc[:4,0:2] # first 4 rows and 2 columns
```

Buscar un dato específico.

#muestra el dato alojado en la fila 1 columna 3


```python
df_books.iloc[1:3] # rows 1 and 2 and all columns
```
