**Leer archivos CSV y JSON con Pandas**

```python
pd.read_csv("data_example.csv")# we use pd.read_csv to read a csv file
```

En algunas ocasiones el archivo podría estar separado por " | " y se vería así.


```python
pd.read_csv("data_example.csv", sep=",") # we can specify the separator, in some cases it is not a comma
```

Cambiar el encabezado, lo podemos hacer con “Header”, este pondrá de encabezado los valores que tenga en esa posición.

```python
pd.read_csv("data_example.csv", header=2) # we can specify the header, in this case the header is in the third row. 
```

* Cambiar el nombre de las columnas con “names”.


```python
pd.read_csv("data_example.csv", header=2, names=["a", "b", "c", "d", "e"])# we can specify the names of the columns using the names parameter
```
