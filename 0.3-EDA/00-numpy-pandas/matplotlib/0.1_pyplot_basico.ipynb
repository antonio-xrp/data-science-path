Pyplot es una herramienta que tiene Matplotlib para ejecutar gráficas de manera sencilla. Véamos cómo puedes lograrlo.


**¿Cómo empezar a usar Pyplot?**

* Importamos Matplotlib y Pandas

```python
import matplotlib.pyplot as plt
import numpy as np
```

* Creamos nuestras variables


```python
x=np.linspace(0,5,11) # we use linspace to create an array of 11 elements from 0 to 5
x
```

```python
y=x**2 # we use the array x to create a new array y with the square of each element of x
y
```

* Graficar en función de y


```python
plt.plot(x,y) # we use the plot function to plot the array x against the array y
plt.show() # we use the show function to display the plot
```

## **Modificar gráficas usando Pyplot**

gráfico de linea

Ahora, vamos a cambiar color de la gráfica

| Character | Color   |
| --------- | ------- |
| 'b'       | Blue    |
| 'g'       | Green   |
| 'm'       | Magenta |

```python
plt.plot(x,y,'k'); #we use 'm' to change the color of the line to magenta, there are many other options for colors for example 'r' for red, 'g' for green, 'b' for blue, 'c' for cyan, 'y' for yellow, 'k' for black, 'w' for white
```

**Cambiar más parámetros Format Strings**

| Character | Description           |
| --------- | --------------------- |
| '.'       | Point marker          |
| ','       | Pixel marker          |
| 'o'       | Circle marker         |
| 'v'       | Triangle_down marker  |
| '^'       | Triangle_up marker    |
| '<'       | Triangle_left marker  |
| '>'       | Triangle_right marker |


**Line Styles**

| Character | Description          |
| --------- | -------------------- |
| '-'       | Solid line style      |
| '–'       | Dashed line style     |
| '-.'      | Dash-dot line style   |
| ':'       | Dotted line style     |

```python
plt.plot(x,y,"yD:"); # we use 'D' to change the marker to a diamond, there are many other options for markers for example 'o' for circle, 's' for square, '^' for triangle, 'p' for pentagon, 'h' for hexagon, 'D' for diamond, 'x' for x, '+' for plus, '*' for star.

# we use ':' to change the line style to a dotted line, there are many other options for line styles for example '-' for solid line, '--' for dashed line, '-.' for dash-dot line, ':' for dotted line.

# we use "yD:" to change the color to yellow, the marker to a diamond and the line style to a dotted line
```

**Crear histograma**

```python
plt.hist(y); # we use the hist function to create a histogram of the array y
```

**Crear gráfica de pie**


```python
plt.pie(y) # we use the pie function to create a pie chart of the array y
plt.show()
```

**Conocer correlación entre variables**


```python
plt.scatter(x,y); # we use the scatter function to create a scatter plot of the array x against the array y

#scatter is a great way to visualize the relationship between two variables, it is also a great way to visualize the distribution of a single variable
```

**Distribución de los datos**

* Distribución de los datos: Gráfico de caja


```python
plt.boxplot(x); # we use the boxplot function to create a boxplot of the array x

#boxplot is a great way to visualize the distribution of a single variable, it indicates the median, the first quartile, the third quartile, the minimum and the maximum
```
