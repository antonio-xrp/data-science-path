# Linux 

### 1. Unificación de sistema de archivos 
1. / : Raíz del sistema de archivos
---
**Estructura de directorios**

2. /bin : Binarios del sistema
3. /boot : Archivos de arranque
4. /dev : Contiene archivos especiales que se utilizan para representar los distintos dispositivos de hardware instalados en el sistema
5. /etc : Archivos de configuración global
6. /home : Directorios de usuarios 
7. /lib : Bibliotecas del sistema
8. /media : Puntos de montaje de dispositivos extraíbles
9. /mnt : Puntos de montaje de dispositivos
10. /opt : Instalaciones de software de terceros
11. /proc : Información del sistema de procesos que se estan ejecutando
12. /root : Directorio del usuario root
13. /run : Archivos temporales
14. /sbin : Binarios del sistema
15. /srv : Datos de servicios
16. /sys : Información del sistema
17. /tmp : Archivos temporales creados por aplicaciones de usuario
18. /usr : Programas y archivos de usuarios
19. /var : Archivos variables, incluidos los archivos de registro del sistema



### 2. Shell

Es un intérprete de comandos que permite al usuario interactuar con el sistema operativo.



### 3. Comandos básicos



1. **Abrir terminal** : Ctrl + Alt + T
2. **Limpiar terminal** : clear
3. **Salir de la terminal** : exit
4. **Mostrar la ruta actual** : pwd
5. **Listar archivos y directorios** : ls
    - ls -l (-l : lista): Muestra los archivos y directorios en forma de lista, en la que se incluye información como permisos, propietario, grupo de usuario al que pertenece, tamaño, fecha y hora de creación, nombre del archivo o directorio.
    
    

    - ls -a (-a : all): Muestra todos los archivos y directorios, incluidos los ocultos(inician con un punto).
    

    - ls -1 (-1 : one): Muestra los archivos y directorios en una sola columna.
    - ls -a -l : Muestra todos los archivos y directorios en forma de lista.
    - ls -l -h ó ls -lh: Muestra los archivos y directorios en forma de lista, con tamaños de archivo legibles por humanos.
    - ls -l nombre_directorio : Muestra los archivos y directorios de un directorio en específico. (Colores azules: Directorio, los colores moras: archivos de enlace simbólico, apuntan a otro archivo o directorio, el texto resaltado: Otros usuarios pueden escribir en ese directorio, archivos sin color: archivos regulares, archivos de texto)
    
     - ls -la : Muestra todos los archivos y directorios en forma de lista, incluidos los ocultos.
    

> Una ruta absoluta es aquella que comienza desde la raíz del sistema de archivos, mientras que una ruta relativa es aquella que comienza desde el directorio actual.

6. **Cambiar de directorio** : cd
    - cd /home : Cambia al directorio home.
    - cd / : Cambia al directorio raíz.
    - cd .. : Cambia al directorio padre.
    - cd ../../ : Cambia al directorio padre del directorio padre.
    - . : Directorio actual.
    - .. : Directorio padre.
    - cd ~ : Te lleva al directorio de usuario.
    - cd - : Te lleva al directorio anterior.

7. **Cambiar marcas de tiempo** : touch
    - touch archivo.txt : Cambia la marca de tiempo de un archivo a la hora actual.
    - touch -t 202107011200 archivo.txt : Cambia la marca de tiempo de un archivo a una fecha y hora específica.
    - touch -d "2021-07-01 12:00" archivo.txt : Cambia la marca de tiempo de un archivo a una fecha y hora específica.
    - touch archivo.txt: Crear un archivo vacío.

8. **Crear directorios** : mkdir
    - mkdir -p directorio1/directorio2/directorio3 : Crea directorios anidados.
    - mkdir -p folder1/{folder2,folder3,folder4} : Crea varios directorios.
    - ls -R : Muestra los directorios y subdirectorios.
9. **Eliminar directorios** : rmdir solo si estan vacios
10. **Eliminar archivos** : rm  (borra archivos del sistema y no se pueden recuperar)
    - rm -r directorio : Elimina un directorio y su contenido.
11. **Copiar archivos** : cp
    - cp archivo1 archivo2 : Copia un archivo a otro archivo.
    - cp archivo1 dir/archivo2 : Copia un archivo a otro archivo cambiandle el nombre.
    - 
    - cp archivo1 archivo2 directorio : Copia un archivo a un directorio.
    - cp -r directorio1 directorio2 : Copia un directorio y su contenido.
