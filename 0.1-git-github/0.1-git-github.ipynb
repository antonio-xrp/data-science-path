# Git y Github

## Introducción

Git es un software de control de versiones distribuido y descentralizado que permite a un equipo de desarrolladores trabajar sobre el mismo código. 

### Configuración Inicial
1. Para ver la versión de git
```git
git --version
```
2. Asignación de tu `user.name` de manera global `--global`
```git
git config --global user.name "Antonio"
```
3. Asignación de tu `user.email` de manera global, especificamente de tu `github`.
```git
git config --global user.email antoniopalaciosorihuela@gmail.com
```
4. Para activar los colores en tu configuracion de git
```git
git config --global user.ui true
```
5. Para observar las configuraciones que tienes en git y `q` para salir
```git
git config --list
```
6. Para asignar vscode como editor de configuración de git.
```git
git config --global core.editor "code --wait"
```
7. Para realizar una edición a la configuracion global desde tu editor de codigo
```git
git config --global -e
```
8. Para estandarizar los saltos de línea en windows
```git
git config --global core.autocrlf true
``` 
9. Para estandarizar los saltos de línea en linux/mac
```git
git config --global core.autocrlf input
``` 
10. Para abrir ayuda de git en la terminal
```git
git config -h
``` 
11. Para abrir ayuda de git completa
```git
git help config
``` 


## Inicializar _Git_ en un directorio local
1. Para iniciar git en un directorio
```git
git init
```

## Flujo básico

El flujo de git consta de 4 estados, dichos estados son: 
1. **_Working Directory_**: Es el área correspondiente al estado _modified_ y es la carpeta local de tu computadora donde almacenas los archivos de tu proyecto.
1. **_Stagin Area_**: Es el área correspondiente al estado _staged_ también se le llama _index_ porque es el área donde _git_ indexa y agrega los cambios realizados en los archivos previos a comprometerlos en su registro.
1. **_Local Repository_**: Es el área correspondiente al estado _comitted_, donde los cambios ya se han registrado de _git_ también llamado _HEAD_ por que indica en que cambio se encuentra el puntero del repositorio. 
1.**_Remote Repository_**: Es el área correspondiente al estado _remote_ y es el directorio remoto donde almacenamos los archivos del proyecto en alguna plataforma _web_ como _github_. `git` denomia **_origin_** al repositorio remoto. 



```git
```
1. Agregar los cambios de un archivo al _staged_
```git
git add archivo/directorio
```
2. Agregar todos los cambios de todos los archivos al staged
```git
git add .
```
---
```git
# Los cambios son comprometidos en el repositorio
# debes escribir el mensaje en el repositorio
# cuando se abre el archivo de configuracion
# al terminar guarda y cierra el archivo ctrl + s
# para los cambios tengan efecto
```
---
3. Para agregar los cambios
```git
git commit 
```
4. Una buena práctica para guardar el commit con un mensaje preciso.
```git
git commit -m "mensaje descriptivo del cambio"
```
5. Agregar el origen remoto de tu repositorio de Github
```git
git remote add origin https://github.com/usuario/repositorio.git
```
6. La primera vez que vinculamos el repositorio remoto con el local
```git
git push -u origin master
```
7. Para las subsecuentes actualizaciones, sino cambias de rama
```git
git push
```
---

8. Para descargar los cambios del repositorio remoto al local de la misma rama.
```
git pull
```
